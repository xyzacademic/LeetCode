```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        # create a new ListNode
        new_head = ListNode()
        tmp = new_head
        
        # quit while loop if any list is empty
        while l1 and l2:
            if l1.val <= l2.val:
                tmp.next = ListNode(l1.val)
                l1 = l1.next 
            else:
                tmp.next = ListNode(l2.val)
                l2 = l2.next 
            tmp = tmp.next
        
        # concatenate to the rest one
        tmp.next = l1 if l1 else l2 
    
        return new_head.next
```
