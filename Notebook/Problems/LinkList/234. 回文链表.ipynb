关键点
=
1. 找到中点（快慢指针-1倍速，2倍速）
2. 翻转后半段链表
3. 遍历比较

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def isPalindrome(self, head: ListNode) -> bool:

        def find_mid_node(head):
            """
            12345 return 3
            1234  return 2
            
            """
            slow = head
            fast = head 

            while fast.next and fast.next.next:
                fast = fast.next.next 
                slow = slow.next 
            
            return slow

        def reverse_list(head):
            cur = None 
            while head:
                tmp = head.next 
                head.next = cur 
                cur = head 
                head = tmp 
            
            return cur 


        mid_node = find_mid_node(head)
        last_half = reverse_list(mid_node.next)

        left = head 
        right = last_half 
        is_palindrome = True 

        # check if it is palindrome
        while right:
            if left.val != right.val:
                is_palindrome = False
                break
            right = right.next 
            left = left.next
            
        # recover link list
        mid_node.next = reverse_list(last_half)

        return is_palindrome
```
